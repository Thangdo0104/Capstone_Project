## Task3: Reporting

### Task3.1: What was the best month for sales? How much was earned that month?

```python
df['Quantity Ordered']=df['Quantity Ordered'].astype('int')
df['Price Each']=df['Price Each'].astype('float')
df['Order ID']=df['Order ID'].astype('int')
df['Month']=df['Month'].astype('int')
df.dtypes
```

```python
df['Sales']=df['Quantity Ordered']*df['Price Each']
df
```

Now moving the Sales column:

```python
moving_column=df.pop('Sales')
df.insert(4,'Sale',moving_column)
df
```

```python
sales_value=df.groupby('Month').sum()['Sale']
sales_value
```

Plot bar chart

```python
months=range(1,13)
plt.figure(figsize=(8,6))
plt.bar(x=months, height=sales_value, width=0.8,align='center')
plt.xticks(months)
plt.xlabel('Month')
plt.ylabel('Sales in USD')
plt.show()
```
